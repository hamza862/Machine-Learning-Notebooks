```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.metrics import accuracy_score, confusion_matrix, roc_auc_score, roc_curve
from pandas_profiling import ProfileReport
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/aniruddhachoudhury/Red-Wine-Quality/master/winequality-red.csv')
```

```python
df
```

```python
pf = ProfileReport(df)
pf.to_widgets()
```

```python
#Initializing the features and label

X = df.drop(columns = 'quality')
Y= df['quality']
```

```python
X
```

```python
Y
```

```python
# Dividing the training and testing data

x_train, x_test, y_train, y_test = train_test_split(X,Y, test_size = 0.2, random_state = 300)
```

```python
# Building a model

dt_model = DecisionTreeClassifier()
dt_model.fit(x_train, y_train)
```

```python
# Let's do some prediction

dt_model.predict(x_test)
```

```python
# Accuracy of the model

dt_model.score(x_test,y_test)
```

```python
# Building a model with different criterion

dt_model_en = DecisionTreeClassifier(criterion='entropy')
dt_model_en.fit(x_train, y_train)
```

```python
# Accuracy of second model

dt_model_en.score(x_test, y_test)
```

```python
dot_for_dt_en = open('dt_model_en_meta.dot', 'w')
tree.export_graphviz(dt_model_en, out_file = dot_for_dt_en, feature_names = X.columns)
```
